```python
# plot asymmetries

# 3701 1.6K
data1=pd.read_csv('out-zn0-data-pack16.dat',nrows=80,usecols=[0,1,2])
data1 = np.array(data1)
fit1=pd.read_csv('out-zn0-data-pack16.dat',usecols=[3,4])
fit1 = np.array(fit1)

# 3703 19.5K
data2=pd.read_csv('out-zn0-data-pack16.dat',nrows=80,usecols=[5,6,7])
data2 = np.array(data2)
fit2=pd.read_csv('out-zn0-data-pack16.dat',usecols=[8,9])
fit2 = np.array(fit2)

# 3706 23.0K
data3=pd.read_csv('out-zn0-data-pack16.dat',nrows=80,usecols=[10,11,12])
data3 = np.array(data3)
fit3=pd.read_csv('out-zn0-data-pack16.dat',usecols=[13,14])
fit3 = np.array(fit3)

# 37 23.8K
data4=pd.read_csv('out-zn0-data-23p8-pack16.dat',nrows=80,usecols=[0,1,2])
data4 = np.array(data4)
fit4=pd.read_csv('out-zn0-data-23p8-pack16.dat',nrows=640,usecols=[3,4])
fit4 = np.array(fit4)


# 3708 25.1K
data5=pd.read_csv('out-zn0-data-pack16.dat',nrows=80,usecols=[15,16,17])
data5 = np.array(data5)
fit5=pd.read_csv('out-zn0-data-pack16.dat',nrows=640,usecols=[18,19])
fit5 = np.array(fit5)
```

```python
# Nice spectra figures

a= 'C:\\Windows\\Fonts\\times.TTF'
fprop = fm.FontProperties(fname=a)
plt.rcParams['font.family'] = fprop.get_name()
plt.rcParams['figure.figsize'] = [8,5]

plt.rc('xtick', labelsize=19)
plt.rc('ytick', labelsize=19)

fig, ax = plt.subplots()

ax.errorbar(data1[:,0],data1[:,1]/data1[0,1],yerr=data1[:,2]/data1[0,1],linestyle='',color = color_map(0, alpha = 0.5), marker = 'D', markersize = 7.5,label='T = 1.6 K')
plt.plot(fit1[:,0],fit1[:,1]/data1[0,1],label='',color=color_map(0, alpha=1), linewidth = 2.2)
ax.errorbar(data2[:,0],data2[:,1]/data2[0,1],yerr=data2[:,2]/data2[0,1],linestyle='',color = color_map(1, alpha = 0.5), marker = 'D', markersize = 7.5,label='T = 19.5 K')
plt.plot(fit2[:,0],fit2[:,1]/data2[0,1],label='',color=color_map(1, alpha=1), linewidth = 2.2)
ax.errorbar(data3[:,0],data3[:,1]/data3[0,1],yerr=data3[:,2]/data3[0,1],linestyle='',color = color_map(2, alpha = 0.5), marker = 's', markersize = 8,label='T = 23.0 K')
plt.plot(fit3[:,0],fit3[:,1]/data3[0,1],label='',color=color_map(2, alpha=1), linewidth = 2.2)
ax.errorbar(data4[:,0],data4[:,1]/data4[0,1],yerr=data4[:,2]/data4[0,1],linestyle='',color = color_map(3, alpha = 0.5), marker = 'o', markersize = 8,label='T = 23.8 K')
plt.plot(fit4[:,0],fit4[:,1]/data4[0,1],label='',color=color_map(3, alpha=1), linewidth = 2.2)
ax.errorbar(data5[:,0],data5[:,1]/data5[0,1],yerr=data5[:,2]/data5[0,1],linestyle='',color = color_map(4, alpha = 0.5), marker = '^', markersize = 9,label='T = 25.1 K')
plt.plot(fit5[:,0],fit5[:,1]/data5[0,1],label='',color=color_map(4, alpha=1), linewidth = 2.2)


ax.set_xlabel('Time ($\mathregular{\mu}$s)',fontsize=20,weight='bold')
ax.set_ylabel('Polarization',fontsize=20,weight='bold')
ax.set_xlim([0,0.5])
ax.set_ylim([0,1.18])
#plt.figtext(0.82,0.2,'Zn0, ZF',fontsize=20)


fig.tight_layout()
ax.legend(loc='upper right',fontsize=14)
plt.savefig('Final_muSR_Zn0_ZF_polarization_Tspectra.png')
plt.show()
```
