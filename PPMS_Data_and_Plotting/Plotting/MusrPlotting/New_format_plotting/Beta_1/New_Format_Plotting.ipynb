```python
from PIL import Image
import os
import math

output_dir = f'{datafile} parameter_plots'

# Get all .png files in the folder (sorted by filename)
plot_filenames = sorted([
    os.path.join(output_dir, f)
    for f in os.listdir(output_dir)
    if f.endswith(".png")
])

# Load images
images = [Image.open(f) for f in plot_filenames]

# Limit to first 9 images (if more)
images = images[:9]

# Grid size
cols = 3
rows = math.ceil(len(images) / cols)

# Resize all to same dimensions (based on max width/height)
max_width = max(img.width for img in images)
max_height = max(img.height for img in images)

# Create blank canvas
combined_width = cols * max_width
combined_height = rows * max_height
combined = Image.new("RGB", (combined_width, combined_height), color=(255, 255, 255))

# Paste images in grid
for index, img in enumerate(images):
    row = index // cols
    col = index % cols
    # Pad image to max size if needed
    padded = Image.new("RGB", (max_width, max_height), (255, 255, 255))
    padded.paste(img, ((max_width - img.width) // 2, (max_height - img.height) // 2))
    combined.paste(padded, (col * max_width, row * max_height))

# Save final combined image
combined_path = os.path.join(output_dir, "combined_plot_grid.png")
combined.save(combined_path)
```

```python
import numpy as np

def muon_asymmetry(
    t,
    f_F,             # ordered volume fraction
    sigma,             # sigma of YBO ordered gaussian
    nu,              # frequency 
    phi,             # phase  in degrees
    lam_tail,        # _tail
    lam_para,        # _paramagnetic
    lam_Ag,# _Ag       
    Background=0,
    beta_tail=1,       # _tail
    beta_para=1,          # _2
    beta_Ag=1,
    A_T=0.2392,
    Ag_F=0.5
    ):
    # Convert phase to radians
    phi_rad = np.pi * phi / 180
    
    #convert field back to frequency
    mu_gyro = 0.0135538817
    nu = nu*mu_gyro
    print(nu)

    # Ordered YBO component 
    term1 = f_F * (2/3) * np.exp(-0.5 * (sigma * t)**2) * np.cos(2 * np.pi * nu * t + phi_rad)
    
    # Ordered Tail exponential decay
    term2 = f_F * (1/3) * np.exp(-(lam_tail * t)**beta_tail)
    
    # Disordered YBO paramagnetic component
    term3 = (1 - f_F) * np.exp(-(lam_para * t)**beta_para)
    
    # Silver (Ag) component
    term4 = np.exp(-(lam_Ag * t)**beta_Ag)
    
    # Full asymmetry function
    A = A_T * (1 - Ag_F) * (term1 + term2 + term3) + A_T * Ag_F * term4 + A_T * Background

    return A
```
