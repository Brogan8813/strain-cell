```python

#Plotting section using defined functions


import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('datafile_parsed.csv')

# Plot dataT vs YBO_L2 for all runs
plt.scatter(df['dataT'], df['Order_Vfrac_'])
plt.xlabel('Temperature (K)')
plt.ylabel('Order volume fraction')
plt.errorbar(df['dataT'], df['Order_Vfrac_'], yerr=df['Order_Vfrac_err'], fmt="o")
plt.ylim(-0.1,1.1)
plt.title('YBO Musr fit parameters')
plt.show()
```

```python

#Plotting section using defined functions


import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('datafile_parsed.csv')

# Plot dataT vs YBO_L2 for all runs
plt.scatter(df['dataT'], df['YBO_L2'], marker = 's', c='r', label = 'YBO_L2')
plt.errorbar(df['dataT'], df['YBO_L2'], yerr=df['YBO_L2_err'], marker = 'x', c='r', fmt="o")

plt.scatter(df['dataT'], df['YBO_B2'], marker = 's', c='b', label = 'YBO_B2')
plt.errorbar(df['dataT'], df['YBO_B2'], yerr=df['YBO_B2_err'], marker = 'o', c='b',fmt="o")
plt.xlabel('Temperature (K)')
plt.ylabel('y')
[plt.axhline(y=i, color='b') for i in [0.4,0.9]]
[plt.axhline(y=i, color='r') for i in [1,3]]

#plt.hlines(0.4,0.9,c='b')
#plt.hlines(1,3,c='r')
plt.legend()
plt.xlim(0.4, 2.3)
plt.ylim(-0.3,3.5)
plt.title('YBO Musr fit parameters')
plt.show()
```
